```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats
```

```python
url=u'https://zenodo.org/record/1171876/files/we_july_2016_may_2017.csv'
```

```python
data = pd.read_csv(url,sep=';')
```

```python
data
```

```python
#df[df['first_name'].notnull() & (df['nationality'] == "USA")]
datap=data[data['WindSpeed']<20.0]
WS=datap.loc[: , "WindSpeed"]
```

```python
plt.plot(WS)
```
